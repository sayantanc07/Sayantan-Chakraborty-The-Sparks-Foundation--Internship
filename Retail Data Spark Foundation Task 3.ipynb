# Data Visualisation

```python
sns.pairplot(data)
```

```python
sns.boxplot(x=data['Profit'])
```

```python
sns.boxplot(x=data['Discount'])
```

```python
sns.boxplot(x=data['Quantity'])
```

```python
plt.figure(figsize=(6,6))
plt.pie(data['Category'].value_counts(),labels=data['Category'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
import seaborn as sns 
fig,axes = plt.subplots(1,2,figsize=(14,5))
fig.suptitle("Total profit VS sales ")
sns.barplot(data=data.groupby('Sub-Category')['Sales','Profit'].agg(sum),x='Sales',y='Profit',ax=axes[1])
data.groupby('Sub-Category')['Sales','Profit'].agg(sum).plot(kind='bar',ax=axes[0])
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns 
fig,axes = plt.subplots(1,2,figsize=(14,5))
fig.suptitle("Total profit VS Quantity ")
sns.barplot(data=data.groupby('Sub-Category')['Quantity','Profit'].agg(sum),x='Quantity',y='Profit',ax=axes[1])
data.groupby('Sub-Category')['Quantity','Profit'].agg(sum).plot(kind='bar',ax=axes[0])
plt.xticks(rotation=90)
plt.show()
```

```python
fig,axes = plt.subplots(2,2,figsize=(20,10))
fig.suptitle("Distribution plots", fontsize=20)
sns.distplot(data['Sales'],ax=axes[0,0])
sns.distplot(data['Profit'],ax=axes[0,1])
sns.distplot(data['Quantity'],ax=axes[1,0])
sns.distplot(data['Discount'],ax=axes[1,1])
plt.show()
```

```python
fig,ax= plt.subplots(1,1,figsize=(12,7))
sns.countplot(data['Quantity'],hue=data['Region'])
plt.show()
```

```python
fig, axes=plt.subplots(2, 2, figsize=(20,10))
fig.suptitle('Visualising categorical columns')
sns.barplot(x=data['Ship Mode'], y=data['Sales'], ax=axes[0,0])
sns.lineplot(x=data['Ship Mode'], y=data['Sales'], ax=axes[0,1])
sns.barplot(x=data['Segment'], y=data['Sales'], ax=axes[1,0])
sns.lineplot(x=data['Segment'], y=data['Sales'], ax=axes[1,1])
plt.show()
```
